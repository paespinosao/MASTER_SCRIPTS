```python
date_ini = datetime(2021,10,21,0) 
date_fin = datetime(2021,11,14,22)
minutes_swan =int((date_fin-date_ini).days*24 + (date_fin-date_ini).seconds/3600)*60 # Minutos totales
time = np.array([date_ini + timedelta(minutes=m) for m in range(0,int(minutes_swan),30)]) # Vector de timepos durante la campa√±a de campo
minutes = np.array([((time[i] - time[0]).days*1440)+((time[i] - time[0]).seconds/60) for i in range (0,len(time))])
zeros = np.zeros(len(time))
df=pd.DataFrame(data={'Time':minutes,'WaterLevel':zeros,'XVeloc':zeros,'YVeloc':zeros})
df
```

```python
filename = 'time_points_d3d.txt'
with open (filename, 'w') as f:
    for i in range(len(df)):
        f.write("[TimePoint]\n")
        a = f'{"Time":<21}'
        b = f'{"=":<3}'
        c = f'{df.iloc[i].Time :<2}'
        d = f'{"WaterLevel":<21}'
        ff = f'{df.iloc[i].WaterLevel:<2}'
        g = f'{"XVeloc":<21}'
        h = f'{df.iloc[i].XVeloc:<2}'
        ii = f'{"YVeloc":<21}'
        j = f'{df.iloc[i].YVeloc:<2}'

        print("   "+a+b+c+'\n')
        print("   "+d+b+ff+'\n')
        print("   "+g+b+h+'\n')
        print("   "+ii+b+j+'\n')
        f.write("   "+a+b+c+'\n')
        f.write("   "+d+b+ff+'\n')
        f.write("   "+g+b+h+'\n')
        f.write("   "+ii+b+j+'\n')
f.close()
```
